```python
#Example 1.3

import numpy as np
import matplotlib.pyplot as plt

T = 288
E_lambda = []

for lamda in range(1, 61):
    E_lambda_i = (3.74 * 10e8) / (lamda**5 * (np.exp(14400 / (lamda * T)) - 1))
    E_lambda.append(E_lambda_i)

lamda_values = list(range(1, 61))

# Plotting
plt.plot(lamda_values, E_lambda)
plt.xlabel('Wavelength [µm]')
plt.ylabel('Intensity [W/m² µm]')
plt.title('Spectral Energy Distribution')
plt.grid(True)
plt.show()
```
